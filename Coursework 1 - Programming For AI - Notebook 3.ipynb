## Task 1: Describe Linear Regression in English
Given a dataset of $n$ values $(x_i, y_i)$ for $1 \leq i \leq n$, describe the process to compute values $\alpha$ and $\beta$ such that $$\sum_{i = 1}^n (y_i - \alpha x_i - \beta)^2$$ is minimal. This will yield a (linear) model $y = \alpha x + \beta$ adhering to the least-squares condition.
> Marking information: Up to 30 points: 15 points for an algorithm that can be followed, and 15 points for the correctness of this algorithm against the specification "linear regression." 

Linear Regression shows a linear relationship between the dependent variable(x) and independent variable(y).

1. Store a list of the xi values from the data set in a variable, data_X
2. Store a list of the yi values from the data set in a variable, data_Y
3. Calculate the sum of xi values from data_X and store in a variable, sum_X
4. Calculate the sum of yi values from data_Y, store in a variable, sum_Y
5. Calculate the sum of mean xi values which is the sum_X / len(data_X) and store in a variable, mean_X
6. Calculate the sum of mean yi values which is the sum_Y / len(data_Y) and store in a variable, mean_Y
7. Calculate and store a list x_deviation for each input value of x. (data_X - mean_X)
8. Calculate and store a list y_deviation for each input value of y. (data_Y - mean_Y)
9. Calculate the sum of (x_deviation * y_deviation) - store in a value, the numerator 
10. Calculate the sum of (x_deviation squared) - store in a value, the denominator 
11. numerator / denominator will give you the beta and we store this as variable, beta
12. mean_Y - beta * mean_X will give you the alpha and we store this as variable, alpha

## Task 3: Implement the Algorithm of Task 1 in Python
Starting from your description of the steps to undertake, transfer English and Maths to Python. Note that you are expected to fully implement the mathematical operations instead of using a library function such as `scikit` or `statsmodels`.
> Marking information: Up to 40 points: 30 points for a correctly working Python implementation, and 10 points for comments and overall readability.

```python
Dataset = {'x': [1, 3, 5], 'y': [8, 10, 15]}

data_X = Dataset['x']
data_Y = Dataset['y']

sum_X = 9
sum_Y = 33

#Formula to work out N

length_X = len(data_X)
length_Y = len(data_Y)

#Total divided by N gives the average 
mean_X = sum_X / length_X
mean_Y = sum_Y / length_Y

#Works out the x deviation by subtracting the x mean from each value of x

x_deviation = []
for data in data_X:
    total = data - mean_X 
    x_deviation.append(total)

#Works out the y deviation by subtracting the y mean from each value of y 
    
y_deviation = []
for data in data_Y:
    total = data - mean_Y 
    y_deviation.append(total)
    
#Numerator for Beta formula 
zipped_lists = zip(x_deviation, y_deviation)
numerator = sum = [x * y for (x, y) in zipped_lists]

num_sum = numerator[0] + numerator[1] + numerator[2]

#Denominator for Beta formula
numerator = [dev**2 for dev in x_deviation]
dom_sum = numerator[0] + numerator[1] + numerator[2]

#Beta Value
beta = num_sum / dom_sum
beta

#Alpha Value 
alpha = mean_Y - (beta)*(mean_X)

print("Beta:", beta)
print("Alpha:", alpha)
```
